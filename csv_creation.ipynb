```python
image = cv2.imread("DATASET/A1/c1/r1/images/e0/c1r1e0n1.tif")
image3 = cv2.imread("DATASET/A1/c3/r1/images/e0/c3r1e0n1.tif")
image4 = cv2.imread("DATASET/A1/c4/r1/images/e0/c4r1e0n1.tif")

# Grayscale to binary

(thresh, blackAndWhiteImage) = cv2.threshold(image, 170, 255, cv2.THRESH_BINARY)
(thresh, blackAndWhiteImage3) = cv2.threshold(image3, 170, 255, cv2.THRESH_BINARY)
(thresh, blackAndWhiteImage4) = cv2.threshold(image4, 170, 255, cv2.THRESH_BINARY)
```
